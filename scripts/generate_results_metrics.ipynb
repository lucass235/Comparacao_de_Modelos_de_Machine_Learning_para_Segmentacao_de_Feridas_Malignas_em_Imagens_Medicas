```python
import pandas as pd

df = pd.read_csv('../data/results_metrics_models.csv')
```

```python
df
```

```python
# remover coluna epoch
df.drop('epoch', axis=1, inplace=True)
df
```

```python
# Defina as cores para as linhas do grfico
colors = ['b', 'g', 'r', 'c']

# Crie um grfico de linha para cada modelo
plt.figure(figsize=(10, 8))
for i, row in df.iterrows():
    plt.plot(df.columns[1:], row[1:], label=row['model'], color=colors[i])

# Adicione rtulos e legenda
plt.xlabel('Mtricas')
plt.ylabel('Valores')
plt.title('Desempenho dos Modelos')
plt.legend()
plt.savefig('../data/imgs/results_metrics_models.png')
# Mostre o grfico
plt.show()

```
